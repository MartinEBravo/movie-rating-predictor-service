```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("alanvourch/tmdb-movies-daily-updates")
```

```python
import os
import pandas as pd

# List files in the downloaded dataset directory
files = os.listdir(path)
print("Files in dataset directory:", files)

# Read the first file
df = pd.read_csv(os.path.join(path, files[0]))

# Print the first 5 rows
df.head()
```

```python
from utils import preprocess

df = preprocess(df)
```

```python
df['production_countries'] = pd.Categorical(df['production_countries'])
df['original_language'] = pd.Categorical(df['original_language'])

# Changing the release_date to release_year
df['release_year'] = pd.to_datetime(df['release_date']).dt.year

df.head()
```

```python
print(df.columns)
```

```python
import hopsworks

project = hopsworks.login(project="dbillLab1")
fs = project.get_feature_store()

tmdb_movie_fg = fs.get_or_create_feature_group(
    name='tmdb_movies',
    description='Movie information including, director, actors, genres, runtime, averagerating',
    version=1,
    primary_key=['id'],
    online_enabled=True,
)
tmdb_movie_fg.insert(df)
```

```python
from utils import split_genres

tmdb_movie_genres_fg = fs.get_or_create_feature_group(
    name='tmdb_movie_genres',
    description='Movie genres',
    version=1,
    primary_key=['id'],
    online_enabled=True,
)

df_genres = split_genres(df)
tmdb_movie_genres_fg.insert(df_genres)
```
